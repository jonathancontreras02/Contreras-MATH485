# 2.2) Define a python function that computes characteristic attributes for services

```python
def get_service_attribute(df):
    
    services = df.groupby('service').agg( 
        
        num_recipients = ('id', 'count'),
        unique_recipients = ('id', 'nunique'), 
        first_date = ('date', 'min'), 
        last_date = ('date', 'max'), 
        num_month = ('month', 'count'), 
        distinct_month = ('month', 'nunique'),
        max_age = ('age', 'max'),
        min_age = ('age', 'min'),
        avg_age = ('age', 'mean'),
        median_age = ('age', 'median')
       
    )
    services['avg_monthly_num_recipients'] = services['unique_recipients']/services['distinct_month']
    

    return services
```

In general, the average age of recipients appears to be in the neighborhood of 30-40 years old. 

The income support service in particular has a very high percentage of recipients that use the service on a monthly basis in comparison to many of the other services. 

I would assume that the number of recipients receiving income supports might increase around the time of tax day which falls in April, let's explore this:

```python
df.groupby(['service', 'month']).agg(
        recipients_per_month=('id', 'nunique')  
    ).loc["Individuals_Receiving_Income_Supports"].sort_values(by='recipients_per_month', ascending=False)
```

We find, however, that the number of recipients using the Individuals_Receiving_Income_Supports service increases as the year goes on. Perhaps this service is not used for things like help with taxes.

```python
df.groupby(['service', 'month']).agg(
        recipients_per_month=('id', 'nunique')  
    ).loc["Homeless_Population"].sort_values(by='recipients_per_month', ascending=False)
```
